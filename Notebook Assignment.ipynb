```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
```

```python
df = pd.read_csv('Data/medical data')
df.head()
```

# VitD_levels

```python
#get values
col='VitD_levels'
mean = df[col].mean()
median = df[col].median()
std = df[col].std()
plus_one_std = mean + std
minus_one_std = mean - std
```

## Graph

```python
fig,ax=plt.subplots(figsize=(10,6))
sns.histplot(x=df[col], kde=True, stat='density')
ax.axvline(mean, color='red', label = f'Mean={mean:,.2f}')
ax.axvline(median, color='green',label = f'Median = {median:,.2f}')
ax.axvline(plus_one_std, color='k',label=f'+1 std = {plus_one_std:,.2f}')
ax.axvline(minus_one_std, color='k', label = f'-1 std = {minus_one_std:,.2f}')
ax.axvspan(minus_one_std, plus_one_std, alpha=.6, color='yellow',zorder=0)
ax.set_title(f'{col} Distribution')
ax.legend();
```

```python
from scipy import stats
kurt = stats.kurtosis(df[col], fisher = False)
kurt
```

## Questions

- Is it Discrete or Continuous?
 - Continuous
- Does it have a skew? If so, which direction (+/-)
 - No
- What type of kurtosis does it display? (Mesokurtic, Leptokurtic, Platykurtic)
 - kurt result around 3 so "mesokurtic"
